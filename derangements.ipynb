# Derangement
In combinatorial mathematics, a derangement is a permutation of the elements of a set in which no element appears in its original position. In other words, a derangement is a permutation that has no fixed points.

```python
SET_SIZE=40 # Set size for which to calculate derangements (complexity of the problem)
```

## Recursion
This algorithm uses recursion to determine the number of derangements. It uses three cases:
- n==1, when the set size is 1, there is not derangement possible.
- n==2, when the set size is 2, there is only one derangement possible. 
- otherwise, n-1 derangements are possible, multiplied by the derangements for (n-1) plus the derangements for (n-2).

This implementation is not efficient for large 𝑛. It recomputes the same values many times.

```python
class CountDerangementsRecursion:
    def __init__(self, set_size):
        self.set_size = set_size

    def count_derangements(self):
        return self.count_derangements_recursion(self.set_size)

    def count_derangements_recursion(self, n):
        if n == 1:
            return 0
        if n == 2:
            return 1
        else:
            return (n - 1) * (self.count_derangements_recursion(n - 1) + self.count_derangements_recursion(n - 2))


for i in range(1, SET_SIZE):
    n = CountDerangementsRecursion(i).count_derangements()
    print("({} -> {})".format(i, n), end=" ")
```

```python
class CountDerangementsTopDown:
    def __init__(self, set_size):
        self.set_size = set_size
        self.sub_solutions = [-1] * (set_size + 1)

    def count_derangements(self):
        return self.count_derangements_rec(self.set_size)

    def count_derangements_rec(self, n):
        if self.sub_solutions[n] != -1:
            return self.sub_solutions[n]
        if n == 1:
            return 0
        if n == 2:
            return 1
        result = (n - 1) * (self.count_derangements_rec(n - 1) + self.count_derangements_rec(n - 2))
        self.sub_solutions[n] = result
        return result


for i in range(1, SET_SIZE):
    n = CountDerangementsTopDown(i).count_derangements()
    print("({} -> {})".format(i, n), end=" ")
```

```python
class CountDerangementsBottomUp:
    def __init__(self, set_size):
        self.set_size = set_size
        self.sub_solutions = [-1] * (set_size + 1)
        for n in range(1, set_size + 1):
            if n == 1:
                self.sub_solutions[n] = 0
            elif n == 2:
                self.sub_solutions[n] = 1
            else:
                self.sub_solutions[n] = (n - 1) * (self.sub_solutions[n - 1] +
                                                   self.sub_solutions[n - 2])

    def count_derangements(self):
        return self.sub_solutions[self.set_size]


for i in range(1, SET_SIZE):
    n = CountDerangementsBottomUp(i).count_derangements()
    print("({} -> {})".format(i, n), end=" ")
```

```python
class CountDerangementsOpt:
    def __init__(self, set_size):
        self.set_size = set_size
        self.solution_n, solution_n_minus_1, solution_n_minus_2 = 0, 0, 0
        for n in range(1, set_size + 1):
            if n == 1:
                self.solution_n = 0
            elif n == 2:
                self.solution_n = 1
            else:
                self.solution_n = (n - 1) * (solution_n_minus_1 +
                                            solution_n_minus_2)
            solution_n_minus_2 = solution_n_minus_1
            solution_n_minus_1 = self.solution_n

    def count_derangements(self):
        return self.solution_n


for i in range(1, SET_SIZE):
    n = CountDerangementsOpt(i).count_derangements()
    print("({} -> {})".format(i, n), end=" ")
```
